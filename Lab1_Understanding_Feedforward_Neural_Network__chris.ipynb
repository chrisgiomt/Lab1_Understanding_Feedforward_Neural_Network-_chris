```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xc = np.linspace(-100, 100, 100000)
yc = xc * (np.sin(xc**2 / 300))

plt.plot(xc, yc)
plt.show()
```

```python
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten

X_train = xc
Y_train = yc
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.6, random_state=42)
```

```python
# model 1
model_1 = Sequential()
model_1.add(Input(shape=(1,)))
model_1.add(Dense(64, activation='relu'))
model_1.add(Dense(64, activation='relu'))
model_1.add(Dense(1))

model_1.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model_1.fit(X_train, y_train, batch_size = 128, epochs = 50)
```

```python
# model 2
# This time I will do a similar model but instead make it have 150 units instead
# of 64. I want to see how much this will improve the model.
model_2 = Sequential()
model_2.add(Input(shape=(1,)))
model_2.add(Dense(150, activation='relu'))
model_2.add(Dense(150, activation='relu'))
model_2.add(Dense(1))

model_2.compile(
    loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model_2.fit(X_train, y_train, batch_size = 128, epochs = 50)
```

```python
# model 3
model_3 = Sequential()
model_3.add(Input(shape=(1,)))
model_3.add(Dense(150, activation='relu'))
model_3.add(Dense(150, activation='relu'))
model_3.add(Dense(150, activation='relu'))
model_3.add(Dense(1))

model_3.compile(
    loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model_3.fit(X_train, y_train, batch_size = 128, epochs = 50)
```
